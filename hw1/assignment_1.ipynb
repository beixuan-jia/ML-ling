```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
# from sklearn.experimental import enable_iterative_imputer
# from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('icnale_data.tab', sep='\t')
df = df.drop(['CEFR'], axis=1)
df = df.replace(r'\+', '', regex=True)
```

```python
# encoder = OneHotEncoder(handle_unknown='ignore')
encoder = LabelEncoder()
```

```python
def gen_df(df, encoder):
    # Select which feature to use in the classifier
    cat_columns = ['country', 'sex', 'area', 'native_teaching']
    num_columns = ['len1', 'len2', 'len3', 'len4', 'self_eval', 'age', 'time_abroad', 'primary_school', 'INTM', 'INSM']

    num_imputer = SimpleImputer(strategy='median')
    cat_imputer = SimpleImputer(strategy='most_frequent')

    # Create two numpy arrays, one for numerical features, one for categorical features which requires ohe later
    num_arr = num_imputer.fit_transform(df[num_columns])
    cat_arr = cat_imputer.fit_transform(df[cat_columns])

    # cat_arr = encoder.fit_transform(cat_arr).toarray()

    frames = []
    for col in cat_arr.T:
        frames.append(encoder.fit_transform(col))
    cat_arr = np.vstack(tuple(frames)).T
    
    
    data_arr = np.hstack((num_arr, cat_arr))

    return data_arr
```

```python
# Split train/dev/test
train_df = df[df['partition'] == 'train']
dev_df = df[df['partition'] == 'dev']
test_df = df[df['partition'] == 'test']

X_train = gen_df(train_df, encoder)
y_train = train_df['vocab_test']
X_dev = gen_df(dev_df, encoder)
y_dev = dev_df['vocab_test']
X_test = gen_df(test_df, encoder)
y_test = test_df['vocab_test']

# #use PCA to make two sets in same dimension
# from sklearn.decomposition import PCA
# pca = PCA(n_components=25)
# X_train = pca.fit_transform(train_arr)
# X_dev = pca.fit_transform(dev_arr)
```
